```python
def pass_or_fail(row, th=10):
    if (row['DV'] >= th and row['BEEE'] >= th and row['FL'] >= th and 
        row['FIMS'] >= th and row['M2'] >= th and row['PP'] >= th):
        return 'Pass'
    else:
        return 'Fail'

df['Pass_Fail'] = df.apply(lambda row: pass_or_fail(row), axis=1)

print(df[['SECTION', 'Pass_Fail']])
```

### Write the code to get how many people are having backlog in each subject for all students(718) using for loop and if condition in pandas,matplotlib. 
### For example:: for(1-718){ if(DV[i]<10) Backlog++; similarly for every subjects M-II, PP, BEEE, FL, FIMS and add to the backlog column.
## After that count of Backlog will show that how many students are having backlogs.

```python
subject=["DV","M2","PP","BEEE","FL","FIMS"]

df["Backlog"] = 0
for i, row in df.iterrows():
    Backlog = 0
    if row["DV"] < 10:
        Backlog+= 1
    if row["M2"] < 10:
        Backlog += 1
    if row["PP"] < 10:
        Backlog += 1
    if row["BEEE"] < 10:
        Backlog+= 1
    if row["FL"] < 10:
        Backlog += 1
    if row["FIMS"] < 10:
        Backlog += 1
    df.at[i, "Backlog"] = Backlog
df
```

### The Backlog column indicates how many subjects a student scored below 10 in. The more subjects a student has with scores under 10, the higher the value of Backlog

```python
df["Backlog_DV"] = 0
df["Backlog_M2"] = 0
df["Backlog_PP"] = 0
df["Backlog_BEEE"] = 0
df["Backlog_FL"] = 0
df["Backlog_FIMS"] = 0
for i, row in df.iterrows():
    df.at[i, "Backlog_DV"] = 1 if row["DV"] < 10 else 0
    df.at[i, "Backlog_M2"] = 1 if row["M2"] < 10 else 0
    df.at[i, "Backlog_PP"] = 1 if row["PP"] < 10 else 0
    df.at[i, "Backlog_BEEE"] = 1 if row["BEEE"] < 10 else 0
    df.at[i, "Backlog_FL"] = 1 if row["FL"] < 10 else 0
    df.at[i, "Backlog_FIMS"] = 1 if row["FIMS"] < 10 else 0
df
```

### The newly added columnsâ€”"Backlog_DV," "Backlog_M2," "Backlog_PP," "Backlog_BEEE," "Backlog_FL," and "Backlog_FIMS"â€”indicate whether the student's score in each subject is below 10. A value of 1 means the student scored less than 10 in that subject, while 0 means the score was 10 or higher

```python
backlog_counts = {
    "DV_Backlogs": (df["DV"] < 10).sum(),
    "M2_Backlogs": (df["M2"] < 10).sum(),
    "PP_Backlogs": (df["PP"] < 10).sum(),
    "BEEE_Backlogs": (df["BEEE"] < 10).sum(),
    "FL_Backlogs": (df["FL"] < 10).sum(),
    "FIMS_Backlogs": (df["FIMS"] < 10).sum()
}

for subject, count in backlog_counts.items():
    print(f"{subject}: {count}")
```

### The output provides a clear summary of the number of students who are facing difficulties in each subject, with the total number of backlogs in each subject displayed. If a subject shows a high backlog count, it may indicate that many students are struggling in that particular subject,

```python
plt.figure(figsize=(10, 5))
melted_df= df[["Backlog_DV", "Backlog_M2", "Backlog_PP", "Backlog_BEEE", "Backlog_FL", "Backlog_FIMS"]].melt(var_name="Subject", value_name="Backlog")
sns.countplot(data=melted_df, x="Subject", hue="Backlog", palette="coolwarm")
plt.title("Backlog Distribution by Subject")
plt.xlabel("Subjects")
plt.ylabel("Count")
plt.legend(title="Backlog", labels=["No Backlog", "Backlog"])
plt.show()
```

### Write a code to get How many students failed in the particular section and also with particular subject

```python
df['PP'] = pd.to_numeric(df['PP'], errors='coerce')
def prog_skills(PP):
    if PP >= 18:
        return "Very Good"
    elif 12 <= PP <= 17:
        return "Good"
    else:
        return "Poor"
    
df['Programming Skills'] = df['PP'].apply(prog_skills)
df
```

```python
df['DV'] = pd.to_numeric(df['DV'], errors='coerce')
def analytical_skills(DV):
    if DV >= 18:
        return "Very Good"
    elif 12 <= DV <= 17:
        return "Good"
    else:
        return "Poor"
    
df['Analytical Skills'] = df['DV'].apply(analytical_skills)
df
```

```python
skill_counts = df['Analytical Skills'].value_counts()
plt.figure(figsize=(8, 8))
sns.set_theme(style="whitegrid")
colors = ['#FF9999', '#66B2FF', '#99FF99', '#FFCC99']  
plt.pie(skill_counts, labels=skill_counts.index, autopct='%1.1f%%', 
        startangle=90, colors=colors, textprops={'fontsize': 12})
plt.title('Distribution of Analytical Skills', fontsize=16)
plt.show()
```
